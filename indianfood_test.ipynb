```python


# Load the custom trained model
model = YOLO('/content/drive/MyDrive/data/model/best_model.pt')
```

```python
import matplotlib.pyplot as plt
import cv2

# Load an image
image_path = '/content/drive/MyDrive/data/train/images/Datacluster Indian Food (1).jpg'
image = cv2.imread('/content/drive/MyDrive/data/train/images/Datacluster Indian Food (1).jpg')

# Run inference
results = model(image)

# Visualize results
plt.imshow(results[0].plot())
plt.axis('off')
plt.show()
```
